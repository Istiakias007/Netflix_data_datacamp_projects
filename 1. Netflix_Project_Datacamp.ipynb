Implementing instruction no 1

1. What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df
```

```python
netflix_df.shape
```

```python
netflix_df_sorted = netflix_df.sort_values(by="release_year", ascending=True)
print(netflix_df_sorted)
```

```python
netflix_df_sorted_decade=netflix_df_sorted[((netflix_df_sorted["release_year"]<=1999) & (netflix_df_sorted["release_year"]>=1990))]
netflix_df_sorted_decade
```

```python
plt.hist(netflix_df_sorted_decade["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
duration_float=netflix_df_sorted_decade["duration"].mode()
duration_int=int(duration_float)
duration=duration_int
print(duration)
```

Implementing 2nd instruction-

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count

```python
netflix_action_movie= netflix_df_sorted_decade[(netflix_df_sorted['type']=='Movie') & (netflix_df_sorted['genre']=='Action')]
netflix_action_movie
```

```python
netflix_short_movie=netflix_action_movie[netflix_action_movie["duration"]<90]
netflix_short_movie.shape
```

```python
short_movie_count=netflix_short_movie.shape[0]
print(short_movie_count)
```
